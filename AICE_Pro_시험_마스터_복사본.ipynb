### **1. 타이타닉 승객 데이터를 이용한 <span style="color:darkgreen">승객 생존 여부</span> 예측 문제**

**[ 데이터 컬럼 설명 ]**

- Survived : 승객이 생존(Y)했는지 사망(N)했는지 여부

- Parch : 부모/자녀의 수
- Ticket : 티켓 번호
- Fare : 팈켓에 대해 지불한 금액
- Cabin : 객실 카테고리
- Embarked : 승객이 탑승한 항구 (C=Cherbourg, Q=Queentown, S=Southampton)

```python
model2.compile(loss = 'sparse_categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])
model2.summary()
```

```python
es = EarlyStopping(monitor='val_loss', patience=3, verbose=1)
checkpoint_path = 'tmp_checkpoint.h5'
cp = ModelCheckpoint(checkpoint_path, monitor='val_loss', verbose=1, save_best_only=True)

history = model2.fit(x_train_pad, y_train_2, epochs=50, batch_size=512,
                      validation_split=0.2, verbose =1, callbacks=[es, cp])
```

### **3. 이미지 데이터를 이용한 <span style="color:darkgreen">코끼리 종류</span> 예측 문제**
- 코끼리 이미지 데이터를 분석 및 전처리한 후 머신러닝 혹은 딥러닝으로 코끼리 종류를 예측하고 결과를 분석하세요.
---

**[ 데이터 설명 ]**
- train 데이터
1. /images/train/African : 아프리카 코끼리 이미지셋
2. /images/train/Asian   : 아시아 코끼리 이미지셋
- test 데이터
1. /images/test


**[ 문제 유의 사항 ]**
- train 데이터셋의 이미지를 이용하여 아프리카/아시아 코끼리 데이터에 대한 학습을 진행하여 예측 모델을 생성한다.
- 위에서 생성된 모델을 활용하여 test 데이터셋의 이미지들에 대해 아프리카 코끼리인지 아시아 코끼리인지에 대한 분류를 진행한다.
- 최종 결과 저장 파일명은 사번_문제번호.csv 로 저장한다. 컬럼1: Filename, 컬럼2: Result
- Result 분류 : African / Asian
- 예측된 데이터의 정확도는 70 % 이상 되어야한다.
- 문제풀이를 위한 셀은 제한없이 생성 가능하다.
- 최종 모델의 저장은 pkl, h5 등 형식은 상관 없고 파일명은 '사번_문제번호.pkl(h5)' 와 같은 형식으로 저정한다.

```python
train_dir = ('/content/drive/MyDrive/AICE/images/train/')
test_dir = ('/content/drive/MyDrive/AICE/images/test/')
```

```python
import tensorflow as tf

train_ds3 = tf.keras.utils.image_dataset_from_directory(
  train_dir,
  label_mode='binary',
  validation_split=0.2,
  subset="training",
  seed=42,
  shuffle=True,
  image_size=(224, 224),
  batch_size=32)

val_ds3 = tf.keras.utils.image_dataset_from_directory(
  train_dir,
  label_mode='binary',
  validation_split=0.2,
  subset="validation",
  seed=42,
  shuffle=True,
  image_size=(224, 224),
  batch_size=32)
```

```python
base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3), weights='imagenet', include_top=False)

base_model.summary()
```

```python
# weight freeze
base_model.trainable = False

inputs = tf.keras.Input(shape=(224, 224, 3))
x = tf.keras.layers.Rescaling(1./127.5, offset=-1)(inputs) # rescaling
x = base_model(x, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)  # 3차원(7, 7, 1280) --> 1차원(1280)으로 줄이기 : GlobalAveragePooling2D
output = tf.keras.layers.Dense(2, activation='softmax')(x)

model3 = tf.keras.Model(inputs=inputs, outputs=output)
model3.summary()
```

```python
model3.compile(optimizer=tf.keras.optimizers.Adam(0.001),  # Optimization
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])             # Metrics / Accuracy
```

```python
# step 3 Callback
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

## EarlyStopping
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=3)

## ModelCheckpoint
checkpoint_path = "my_checkpoint.h5"
checkpoint = ModelCheckpoint(filepath=checkpoint_path,
                             save_best_only=True,
                             monitor='val_loss',
                             verbose=1)
```

```python
# 모델 학습(fit)
history = model3.fit(
    train_ds3,
    validation_data = val_ds3,
    epochs=100,
    callbacks=[es, checkpoint]
)
```

```python
model3.evaluate(val_ds3)
```

```python
import os
import numpy as np

result=[]

from keras.preprocessing import image
file_name=os.listdir(test_dir)

for item in file_name :
  img = image.load_img(test_dir+item, target_size=(224,224))
  x=image.img_to_array(img)
  x=np.expand_dims(x, axis=0)
  preds = np.argmax(model3.predict(x))
  result.append([item,train_ds3.class_names[preds]])
```
